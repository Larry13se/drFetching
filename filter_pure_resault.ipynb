```python
import pandas as pd

def reconcile_csv(result_file, pure_file, output_file, join_column):
    """
    Reconciles two CSV files by removing rows from the pure_file that exist in the result_file.

    Args:
        result_file (str): Path to the result CSV file.
        pure_file (str): Path to the pure CSV file.
        output_file (str): Path to save the reconciled pure CSV file.
        join_column (str): The column to use for matching rows between the two files.
    """
    try:
        # FIX: Added encoding='latin1' to handle non-utf8 characters
        df_result = pd.read_csv(result_file, encoding='latin1')
        df_pure = pd.read_csv(pure_file, encoding='latin1')
    except FileNotFoundError:
        print("Error: One or both of the input files were not found.")
        return
    except Exception as e:
        print(f"Error reading CSV files: {e}")
        return

    # Get the IDs from the result file that need to be removed from the pure file
    # Ensure both join columns are of the same type to prevent matching issues
    df_result[join_column] = df_result[join_column].astype(str)
    df_pure[join_column] = df_pure[join_column].astype(str)
    
    ids_to_remove = df_result[join_column].tolist()

    # Filter out rows from df_pure where the join_column value is in ids_to_remove
    df_reconciled = df_pure[~df_pure[join_column].isin(ids_to_remove)]

    # Save the reconciled DataFrame to a new CSV file
    try:
        # It's good practice to save with utf-8-sig to ensure Excel reads it correctly
        df_reconciled.to_csv(output_file, index=False, encoding='utf-8-sig')
        print(f"Reconciliation complete. Remaining rows saved to {output_file}")
        print(f"Original pure file rows: {len(df_pure)}")
        print(f"Rows removed: {len(df_pure) - len(df_reconciled)}")
        print(f"Reconciled pure file rows: {len(df_reconciled)}")
    except Exception as e:
        print(f"Error writing output file: {e}")

if __name__ == "__main__":
    # Example usage:
    result_csv = "final_results.csv"
    pure_csv = "Submission Data - 178 Lead.csv"
    output_csv = "Submission Data - 178 Lead.csv" # It's safer to not overwrite the original input file
    column_to_match = "MEDICARE ID"

    reconcile_csv(result_csv, pure_csv, output_csv, column_to_match)
```
